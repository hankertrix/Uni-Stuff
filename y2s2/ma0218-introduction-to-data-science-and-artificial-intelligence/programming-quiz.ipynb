## Question 1: Exploratory analysis

### a)

Read the CSV file.

```python
satisfaction_data = pd.read_csv("Customer_Satisfaction_Data.csv")
satisfaction_data.head()
```

Remove all occurrences of UNKNOWN.

```python
# Coerce everything to numeric data
cleaned_satisfaction_data = satisfaction_data.apply(lambda elem: pd.to_numeric(elem, errors="coerce"))

# Remove all the infinite values
cleaned_satisfaction_data = cleaned_satisfaction_data.replace([np.inf, -np.inf], np.nan)

# Drop all NaN values
cleaned_satisfaction_data.dropna(axis=0)
```

### b)

Check the information of the data set.

```python
cleaned_satisfaction_data.info()
```

Plot the 3 statistical distributions for the continuous variables.

```python
continuous_variables = ["Order_Total_Amount", "Product_Quality_Rating", "Customer_Age", "Customer_Service_Rating", "Delivery_Time", "Product_Price"]

for variable in continuous_variables:

    # Plot a box plot
    plt.figure(figsize=(16,8))
    sb.boxplot(data=cleaned_satisfaction_data[variable], orient="h")

    # Plot a histogram plot
    plt.figure(figsize=(16,8))
    sb.histplot(data=cleaned_satisfaction_data[variable])

    # Plot a violin plot
    plt.figure(figsize=(16,8))
    sb.violinplot(data=cleaned_satisfaction_data[variable])
```

Plot the pair plots for all the variables.

```python
sb.pairplot(cleaned_satisfaction_data)
```

Based on the pair plot above, it looks like customer service rating, delivery time and product quality rating are the most useful in predicting customer satisfaction.

## Question 2: Uni-Variate Decision Tree

### a)

Read the training and test data.

```python
train_data = pd.read_csv("Customer_train.csv")
test_data = pd.read_csv("Customer_test.csv")
```

Fit the features into a decision tree.

```python
# Import the decision tree classifier from sci-kit learn
from sklearn.tree import DecisionTreeClassifier

# The features above
features = ["Customer_Service_Rating", "Delivery_Time", "Product_Quality_Rating"]

# The list of models
models = []

# Iterate over the features
for feature in features:

    # Initialise the decision tree classifier with maximum depth of 3
    model = DecisionTreeClassifier(max_depth=3)

    # Fit the decision tree classifier
    model.fit(pd.DataFrame(train_data[feature]), pd.DataFrame(train_data["Customer_Satisfaction"]))

    # Append the fitted model to the list of models
    models.append(model)
```

### b)

Predict the customer satisfaction using each of the models.

```python
# Import the confusion matrix function from scikit learn
from sklearn.metrics import confusion_matrix

# The list of matrices
matrices = []

# Iterate over the models
for index, model in enumerate(models):

    # Get the prediction of the model for the training data
    train_prediction = model.predict(pd.DataFrame(train_data[features[index]]))

    # Get the confusion matrix for the training data
    train_matrix = confusion_matrix(pd.DataFrame(train_data["Customer_Satisfaction"]), train_prediction)

    # Print out the confusion matrix for the feature for the training data
    print(f"Train matrix for {features[index]}:")
    print(train_matrix)
    print()

    # Get the prediction of the model for the test data
    test_prediction = model.predict(pd.DataFrame(test_data[features[index]]))

    # Get the confusion matrix for the test data
    test_matrix = confusion_matrix(pd.DataFrame(test_data["Customer_Satisfaction"]), test_prediction)

    # Print out the confusion matrix for the feature for the test data
    print(f"Test matrix for {features[index]}:")
    print(test_matrix)

    # Print a separator
    print("\n\n")

    # Add the training matrix and the test matrix to the list of matrices
    matrices.append((train_matrix, test_matrix))
```

### c)

Print the classification accuracy and the recall scores.

```python
# Iterate over all the models
for index, model in enumerate(models):

    # Print the feature name
    print(features[index])

    # Get the accuracy of the model for the training data
    train_accuracy = model.score(
        pd.DataFrame(train_data[features[index]]),
        pd.DataFrame(train_data["Customer_Satisfaction"])
    )

    # Print out the accuracy of the model for the training data
    print(f"{train_accuracy = }")

    # Get the accuracy of the model for the test data
    test_accuracy = model.score(
        pd.DataFrame(test_data[features[index]]),
        pd.DataFrame(test_data["Customer_Satisfaction"])
    )

    # Print the accuracy of the model for the test data
    print(f"{test_accuracy = }")

    # Get the confusion matrices from the list of matrices
    train_matrix, test_matrix = matrices[index]

    # Get all the values required from the training confusion matrix
    tp_train, fp_train, fn_train, tn_train = train_matrix.ravel()

    # Calculate the recall for the training data
    train_recall = tp_train / (tp_train + fn_train)

    # Get all the values required from the test confusion matrix
    tp_test, fp_test, fn_test, tn_test = test_matrix.ravel()

    # Calculate the recall for the test data
    test_recall = tp_test / (tp_test + fn_test)

    # Print out the recall for both the training and test data
    print(f"{train_recall = }")
    print(f"{test_recall = }")

    # Print a separator
    print()
```

## Clustering patterns

Read the service data.

```python
service_data = pd.read_csv("Service_Rating_and_Delivery_Time.csv")
service_data.head()
```

```python
# Import the K means clustering model from sci-kit learn
from sklearn.cluster import KMeans

# The possible number of clusters
possible_number_of_clusters = list(range(1, 21))

# Initialise the list of the within-cluster sum of squares
within_cluster_sum_of_squares_list = []

# Iterate over the number of possible clusters
for number_of_clusters in range(1, 21):

    # Initialise the model with the number of clusters
    model = KMeans(n_clusters=number_of_clusters)

    # Fit the model to the service data
    model.fit(service_data)

    # Get the centres
    centres = model.cluster_centres_

    # Predict the clusters for the data
    predicted_clusters = model.predict(service_data)

    # Initialise the variable to store the
    # within cluster sum of squares
    within_cluster_sum_of_squares = 0

    # Iterate over the points in the data
    for point_index, point in enumerate(np.array(service_data)):

        # Get the centre for the current point
        point_centre = centres[predicted_clusters[point_index]]

        # Get the x and y coordinates of the point
        point_x, point_y = point

        # Get the x and y coordinates of the point centre
        point_centre_x, point_centre_y = point_centre

        # Calculate the squared Euclidean distance
        # using Pythagoras' theorem, which is
        # sqrt((x_2 - x_1) ** 2 + (y_2 - y_1) ** 2)
        #
        # But since we need the squared Euclidean distance,
        # we can omit the square root, so we get:
        # (x_2 - x_1) ** 2 + (y_2 - y_1) ** 2
        squared_euclidean_distance = (
            (point_x - point_centre_x) ** 2
            + (point_y - point_centre_y) ** 2
        )

        # Add the squared Euclidean distance for the point
        # to the within-cluster sum of squares
        within_cluster_sum_of_squares += squared_euclidean_distance

    # Add the within-cluster sum of squares to the list
    within_cluster_sum_of_squares_list.append(
        within_cluster_sum_of_squares
    )

# Plot the graph of within-cluster sum of squares (y)
# against the number of clusters (x)
plt.figure(figsize=(16, 8))
plt.plot(
    possible_number_of_clusters,
    within_cluster_sum_of_squares,
)

# Force the x-tick labels to display the number of clusters,
# which are integers, instead of the default
plt.xticks(possible_number_of_clusters)

# Label the x and y axis
plt.xlabel("Number of clusters")
plt.ylabel("Within-clutser sum of squares (WSS)")
```
